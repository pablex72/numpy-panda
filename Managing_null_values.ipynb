```python
import pandas as pd
import numpy as np
```

```python
dict = {'Col1' : [1,2,3,np.nan],
        'Col2' : [4,np.nan,6,7],
        'Col3' : ['a','b','c', None]
}
```

```python
df = pd.DataFrame(dict)
```

```python
df.isnull()
```

```python
df.isnull() * 1
```

```python
df.fillna('Missing')
```

```python
df.fillna(df.mean())
```

```python
#hace una interpolacion de la serie asumiendo cual es el valor que deberia seguir
df.interpolate()
```

```python
df.dropna()
```
